```python
import pandas as pd
```

```python
df_movies = pd.read_csv('../dataset/movies.csv')
df_ratings = pd.read_csv('../dataset/ratings.csv')
```

```python
df_movies.head()
```

```python
df_ratings.head()
```

```python
df_movies_data = pd.merge(df_ratings, df_movies, on='movieId')
```

```python
df_movies_data.head()
```

```python
df_movies_data.groupby('title')['rating'].mean().head()
```

```python
df_movies_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df_movies_data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating_mean_count = pd.DataFrame(df_movies_data.groupby('title')['rating'].mean())
rating_mean_count.head()
```

```python
rating_mean_count['rating_counts'] = pd.DataFrame(df_movies_data.groupby('title')['rating'].count())
```

```python
rating_mean_count.head()
```

```python
rating_mean_count['rating'] = round(rating_mean_count['rating'], 1)
```

```python
rating_mean_count['rating'].head()
```

```python
high_rating = rating_mean_count[(rating_mean_count['rating'] > 4) & (rating_mean_count['rating_counts'] > 100)]
high_rating
```

```python
high_rating = high_rating.sort_values(by='rating', ascending=False)
```
