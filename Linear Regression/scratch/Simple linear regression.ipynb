### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

### Load the data

```python
data = pd.read_csv('dataset/Simplelinearregression.csv')
```

```python
data.head()
```

```python
data.describe()
```

### create regression

##### define dependent and independent varaible

```python
y = data['GPA']
x1 = data['SAT']
```

##### Explore the data

```python
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python
sns.regplot(x='SAT', y='GPA', data=data)
plt.show()
```

### Regression

```python
x = sm.add_constant(x1)
result = sm.OLS(y, x).fit()
result.summary()
```

```python
plt.scatter(x1, y)
yhat = 0.0017 * x1 + 0.2750
fig = plt.plot(x1, yhat, lw=4, c='orange', label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.xlabel('GPA', fontsize=20)
plt.legend(loc='lower right')
plt.show()
```
