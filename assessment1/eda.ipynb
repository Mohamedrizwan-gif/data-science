```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset/fifa.csv')
```

```python
df.head()
```

```python
df.drop(['ID', 'Contract Valid Until'], axis=1, inplace=True)
```

#### Missing Data

```python
df.isna()
```

```python
plt.title('Finding Missing values')
sns.heatmap(df.isna(), yticklabels=False, cbar=True, cmap='magma')
plt.show()
```

```python
df.isna().sum()
```

```python
plt.style.use('ggplot')
```

```python
plt.title('Foot preferrence of players')
sns.countplot(x='Preferred Foot', data=df)
plt.show()
```

```python
sns.displot(df['Age'], kde=True)
plt.show()
```

```python
sns.histplot(df['Nationality'][:10])
plt.title('Total Number of football players in a country')
plt.tight_layout()
plt.show()
```

```python
india = df[df['Nationality'] == 'India']
```

```python
sns.heatmap(india.isna(), yticklabels=False)
plt.title('Finding Missing Value')
plt.show()
```

```python
sns.countplot(x=india['Preferred Foot'], data=df)
plt.title('Total Count of preferred foot players in india')
plt.show()
```

```python
sns.countplot(x='Preferred Foot', data=india)
plt.title('Total Count of preferred foot players in india')
plt.show()
```

```python
sns.boxplot(data=india.iloc[:, 0:7])
plt.tight_layout()
plt.show()
```
