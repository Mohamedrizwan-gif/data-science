```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(columns=['Name', 'Gender', 'Maths', 'Science', 'Social', 'English', 'Computer'])
```

```python
df['Name'] = ['Student A','Student B','Student C','Student D','Student E',
              'Student F','Student G','Student H','Student I','Student J']
df['Gender'] = ['Male','Female','Male','Female','Male','Male','Female','Female','Male','Female']
df['Maths'] = np.random.randint(20, 100, 10)
df['Science'] = np.random.randint(20, 100, 10)
df['Social'] = np.random.randint(20, 100, 10)
df['English'] = np.random.randint(20, 100, 10)
df['Computer'] = np.random.randint(20, 100, 10)
df.head(10)
```

#### Create a new column with total of all the marks

```python
df['Total'] = df['Maths'] + df['Science'] + df['Social'] + df['English'] + df['Computer']
```

```python
df
```

#### Rank the students based on the total in a new column

```python
df['Rank'] = df['Total'].rank(ascending=False)
```

```python
df.sort_values(by='Rank', inplace=True)
```

```python
df
```

#### Filter top 3 students

```python
df[df['Rank'] < 4]
```

#### Add a new student with a 100% score in all the subjects and do the ranking again

```python
new_student = {
    'Name': ['Student K'],
    'Gender': ['Male'],
    'Maths': [100],
    'Science': [100],
    'Social': [100],
    'English': [100],
    'Computer': [100],
    'Total': [500]
}

df_new = pd.DataFrame(new_student)
df_new
```

```python
df = pd.concat([df, df_new])
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
df
```

```python
df['Rank'] = df['Total'].rank(ascending=False)
df
```

```python
df.sort_values(by='Rank')
```
