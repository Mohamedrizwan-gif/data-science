```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/fifa.csv')
```

```python
df.head()
```

```python
sns.heatmap(df.isna(), cbar=True, cmap='viridis')
plt.title('Finding null values')
plt.show()
```

```python
plt.title('Box plot')
sns.boxplot(data=df[['Age', 'Overall', 'Club', 'Wage', 'Height', 'Weight']])
plt.show()
```

```python
sns.regplot(x='Age', y='Weight' ,data=df)
plt.title('Age and Height')
plt.tight_layout()
plt.show()
```

```python
sns.lmplot(x='Age', y='Height', data=df)
plt.title('Age and Height')
plt.tight_layout()
plt.show()
```

```python
sns.lmplot(x='Age', y='Height', hue='Preferred Foot', data=df)
plt.title('Age and Height')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True)
plt.tight_layout()
plt.show()
```
