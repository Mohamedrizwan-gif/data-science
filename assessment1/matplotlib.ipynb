```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/fifa.csv')
```

```python
df.head()
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
df.dtypes
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
club = df['Club'].value_counts()
club
```

```python
x = club.index.tolist()
y = club.values.tolist()
```

```python
plt.barh(x[:11], y[:11])
plt.xlabel('total players')
plt.ylabel('club')
plt.title('First 10 most players in the club')
plt.show()
```

```python
plt.barh(x[11:21], y[11:21])
plt.xlabel('total players')
plt.ylabel('club')
plt.title('First 10 most players in the club')
plt.show()
```

```python
plt.barh(x[21:31], y[21:31])
plt.xlabel('total players')
plt.ylabel('club')
plt.title('First 10 most players in the club')
plt.tight_layout()
plt.show()
```

```python
plt.hist(df['Weight'], bins=5)
plt.title('Weight')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
unique_age = df['Age'].unique()
```

```python
plt.plot(unique_age, '--o', label='Age')

plt.ylabel('Age')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.hist(df['Age'])
plt.title('Total Persons in Specific Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.axvline(df['Age'].median(), color='red', label='Age median')
plt.legend()
plt.tight_layout()
plt.show()
```
