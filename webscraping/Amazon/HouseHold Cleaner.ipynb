```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.amazon.in/alm/category/?almBrandId=ctnow&node=1374544031&ref_=fs_rfc_dsk_sf_ai_1374544031_sml_2&pd_rd_w=K5TR6&content-id=amzn1.sym.e8261eb6-2597-478d-a395-00ac3f9522b7&pf_rd_p=e8261eb6-2597-478d-a395-00ac3f9522b7&pf_rd_r=XQ03664RAACWJZ8TNSKZ&pd_rd_wg=bK437&pd_rd_r=460814c7-42ae-499b-9600-8f3796906197#'
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
bs = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
container = bs.findAll('div', 'a-carousel-container')
len(container)
```

```python
col_title = []
col_desc = []
col_new_price = []
col_old_price = []
col_img = []

for ct in container:
    title = ct.find('h2', 'a-size-large').span.text
    product = ct.findAll('li', 'a-carousel-card')
    for pr in product:
        desc = pr.find('span', 'a-truncate')
        img = pr.find('img').get('src')
        if desc:
            new_price = pr.find('span', 'a-text-price').span.text
#             print(desc.text)
#             print(new_price.span)
            try:
                old_price = pr.find('span', 'a-text-strike').text
            except AttributeError:
                old_price = ''
            desc = desc.text
            col_title.append(title)
            col_desc.append(desc)
            col_img.append(img)
            col_new_price.append(new_price)
            col_old_price.append(old_price)
```

```python
data = {
    'title': col_title,
    'description': col_desc,
    'image_url': col_img,
    'new price': col_new_price,
    'old price': col_old_price
}

df = pd.DataFrame(data)
df
```
