```python
import tensorflow as tf

affine_img = tf.keras.preprocessing.image.apply_affine_transform(img, tx=30, ty=30, zx=0.5, zy= 0.5)
plt.imshow(affine_img)
plt.axis('off')
plt.show()
```

```python
from torchvision import transforms
```

```python
affine_img = transforms.RandomAffine((20))
men_img = Image.open('/content/men.jpg')
img = affine_img(men_img)
plt.imshow(img)
plt.show()
```
