```python
df = raw_data.copy()
```

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].astype('category')
```

```python
df['diagnosis'].dtypes
```

```python
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
x = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
from scipy.stats import zscore

xscaled = x.apply(zscore)
xscaled.describe()
```

```python
xscaled_df = pd.DataFrame(xscaled)
```

```python
xscaled_df.head()
```

### Build KNN Model

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(xscaled_df, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=5, weights='distance')
model.fit(xtrain, ytrain)
```

```python
xpredict = model.predict(xtrain)
xpredict
```

```python
ypredict = model.predict(xtest)
ypredict
```

```python
model.score(xtest, ytest)
```

```python
from sklearn import metrics
```

```python
print('Confusion Matrix')

cm = metrics.confusion_matrix(ytest, ypredict, labels=['M', 'B'])
df_cm = pd.DataFrame(cm, index=['M', 'B'], columns=['Predicted M', 'Predicted B'])

plt.figure(figsize=(10, 7))
sns.heatmap(df_cm, annot=True)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = []

for i in range(1, 50):
    model = KNeighborsClassifier(n_neighbors=i)
    scores_ = cross_val_score(model, xscaled_df, y, cv=10)
    scores.append(scores_.mean())
```

```python
len(scores)
```

```python
plt.figure(figsize=(20, 10))

plt.plot(range(1, 50), scores, c='b', ls='--', marker='o', mfc='red', ms=10)
plt.title('Accuracy Rate vs K-values')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.show()
```
